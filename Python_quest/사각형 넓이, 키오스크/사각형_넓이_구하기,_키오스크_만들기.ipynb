```python
class Square:
    def __init__(self):
      self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요. \n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

      match self.square:
        case 1: print('직사각형 함수는 rect()입니다.')   # Square 클래스에 rect() 메서드를 추가합니다.
        case 2: print('평행사변형 함수는 par()입니다.')  # Square 클래스에 par() 메서드를 추가합니다.
        case 3: print('사다리꼴 함수는 trape()입니다.')  # Square 클래스에 trape() 메서드를 추가합니다.
        case _: print('1, 2, 3 중에서 다시 입력해주세요')


    def rect(self):
      """
          직사각형의 넓이 반환
      """

      #가로, 세로 입력
      width, height = map(float, input('가로, 세로를 입력하세요. \
                                        예시 : 1,20.3 \n >>>').split(','))
      return f'직사각형의 넓이는: {width * height}'


    def par(self):
      """
          평행사변형 넓이 반환
      """

      #밑변, 높이 입력
      bottom, height = map(float, input('밑변, 높이를 입력하세요. \
                                        예시 : 3,4.0 \n >>>').split(','))
      return f'평행사변형의 넓이는: {bottom * height}'


    def trape(self):
      """
          사다리꼴 넓이 반환
      """

      # 윗변, 밑변, 높이 입력
      upper, bottom, height = map(float, input('윗변, 밑변, 높이를 입력하세요. \
                                                예시 : 1,2,3 \n >>>').split(','))
      return (upper + bottom) * height / 2
```

```python
a = Square()
b = Square()
c = Square()

print('\n\n')
print(a.rect(), end = '\n\n')
print(b.par(), end = '\n\n')
print(c.trape(), end = '\n\n')
```

```python
class Kiosk:
  def __init__(self):
    self.menu  = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
    self.price = [2000, 3000, 3000, 2500, 2500, 3000]

  # 메뉴 출력 메서드
  def menu_print(self):
    for i in range(len(self.menu)):
      print(i + 1, self.menu[i], ' : ', self.price[i], '원')


  # 주문 메서드
  def menu_select(self):
    self.order_menu  = []    # 주문 리스트
    self.order_price = []    # 가격 리스트

    while True:
      ord_num = int(input("\n음료 번호를 입력하세요 : "))  # 음료 번호 입력

      # 메뉴판에 있는 음료 번호일때
      if 1 <= ord_num <= len(self.menu):
        self.order_price.append(self.price[ord_num-1])    # 가격 리스트에 추가
        self.price_sum = sum(self.order_price)            # 합계 금액
        break

      # 메뉴판에 없는 번호일 때
      else: print("없는 메뉴입니다. 다시 주문해 주세요.\n")


    # 음료 온도 물어보기
    while True:
      temp = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))

      if   temp == 1: self.temp = 'HOT'; break
      elif temp == 2: self.temp = 'ICE'; break
      else          : print("1과 2 중 하나를 입력하세요. 다시 입력해주세요.\n")

    # 주문 리스트에 추가
    self.order_menu.append(self.temp + ' ' + self.menu[ord_num-1])

    # 온도 속성을 추가한 주문 결과 출력
    print('주문 음료', self.temp, self.menu[ord_num-1], ' : ', self.price[ord_num-1], '원\n')


    # 추가 주문 또는 지불
    while True:
      ord_num = int(input("\n추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))  # 추가 주문 또는 지불

      if 1 <= ord_num <= len(self.menu):
        while True:
          temp = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))

          if   temp == 1: self.temp = 'HOT'; break
          elif temp == 2: self.temp = 'ICE'; break
          else          : print("1과 2 중 하나를 입력하세요. 다시 입력해주세요.\n")

        self.order_menu.append(self.temp + ' ' + self.menu[ord_num-1])        # 주문 리스트에 추가
        self.order_price.append(self.price[ord_num-1])                        # 가격 리스트에 추가
        self.price_sum = sum(self.order_price)
        print('[추가 주문 음료] >> ', self.temp, self.menu[ord_num-1], ':',\
                                     self.price[ord_num-1], '원\n', '합계 : ', self.price_sum)
      elif ord_num == 0:
        print("주문이 완료되었습니다.\n")
        print(self.order_menu, self.order_price)   # 확인을 위한 리스트를 출력
        break

      # 유효하지 않은 입력
      else: print("없는 메뉴입니다. 다시 주문해 주세요.")


  # 지불 메서드
  def pay(self):
    while True:
      pay_type = int(input('지불 방식을 입력하세요. (현금 -> 1, 카드 -> 2) >> '))     # 지불 방식 지정

      if   pay_type == 1: print('\n직원을 호출하겠습니다.'); break                     # 현금 지불
      elif pay_type == 2: print('\nIC칩 방향에 맞게 카드를 꽂아주세요.'); break         # 카드 지불
      else              : print('지불 방식은 1과 2 중 하나를 입력해 주세요. 다시 입력해 주세요.'); continue


  # 주문표 출력 메서드
  def table(self):

    # 외곽
    print('⟝' + '-' * 30 + '⟞')
    for i in range(5):
      print('|' + ' ' * 31 + '|')

    # 주문 상품명 : 해당 금액
    for i in range(len(self.order_menu)):
      print(' ' * 5, self.order_menu[i][:9], ' : ', self.order_price[i])

    print('\n' + ' ' * 6 + '[합계 금액] :', self.price_sum)

    # 외곽
    for i in range(5):
      print('|' + ' ' * 31+ '|')
      print('⟝' + '-' * 30 + '⟞')
```

```python
a = Kiosk()
a.menu_print()
a.menu_select()
a.pay()
a.table()
```
