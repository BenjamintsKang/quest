```python
from random import randint
from datetime import datetime

class Account:
  BANK_NAME = 'SC은행'                                   # Q1. 은행이름
  INTEREST_RATE = 1.01                                   # Q7. 이자 설정

  num_of_accounts = 0                                    # Q2. 계좌 생성 개수


  def __init__(self):
    self.__acc_holder = input('[예 금 주] : ')            # Q1
    self.__acc_num    = self.__creat_acc_num()            # Q1
    self.__transac_info = {'입금내역':[], '출금내역':[]}   # Q10
    Account.num_of_accounts += 1                          # Q2

    self.__balance    = 0                                 # Q1
    self.__num_of_deposits = 0                            # Q7

    self.deposit(int(input('[입 금 액] $ ')))              # Q1
    if self.__balance > 0: self.__num_of_deposits = 1
    else: self.__num_of_deposits = 0


  def __creat_acc_num(self):                               # Q1
    unique_nums = (str(randint(100, 999)), str(randint(10, 99)), \
                   str(randint(100000, 999999)))
    return '-'.join(unique_nums)


  @property
  def get_account_num(self):                               # Q3
    print(f'\n[계좌 번호] # {self.__acc_num}\n')


  @property
  def balance(self):
    return self.__balance


  def deposit(self, deposit_amount):                        # Q4
    if deposit_amount >= 1:
      if self.__num_of_deposits == 5:
        self.__num_of_deposits = 0
        # Q7
        self.__balance = round(Account.INTEREST_RATE * (self.__balance + deposit_amount))
        self.__transac_info['입금내역'].append(\
                                           ({'거래일': datetime.today().strftime('%Y-%m-%d %H:%M:%S'),\
                                            '맡기신금액': f'{deposit_amount:,}',\
                                            '남은금액': f'{self.__balance:,}'}))
        print('\n', self.__transac_info['입금내역'][-1])
      else:
        self.__balance += deposit_amount
        self.__transac_info['입금내역'].append(\
                                           ({'거래일': datetime.today().strftime('%Y-%m-%d %H:%M:%S'),\
                                            '맡기신금액': f'{deposit_amount:,}',\
                                            '남은금액': f'{self.__balance:,}'}))
        print('\n', self.__transac_info['입금내역'][-1])
    else: print('고객님, 입금은 최소 1원 이상이 가능합니다.\n\n')


  def withdraw(self, withdrwal_amount):                      # Q5
    if withdrwal_amount <= self.__balance:
      self.__balance -= withdrwal_amount
      self.__transac_info['출금내역'].append(\
                                           ({'거래일': datetime.today().strftime('%Y-%m-%d %H:%M:%S'),\
                                            '찾으신금액': f'{withdrwal_amount:,}',\
                                            '남은금액': f'{self.__balance:,}'}))
      print('\n', self.__transac_info['출금내역'][-1])
    else:
      print('잔고가 부족합니다.')
      print('**** 잔 고 ****   $ {self.balance:,}\n')


  @property
  def display_info(self):                                     # Q6
    print('\n\n')
    print(f'[ 은  행  명  ]   : {Account.BANK_NAME}')
    print(f'[ 예  금  주  ]   : {self.__acc_holder}')
    print(f'[ 계 좌 번 호 ]   # {self.__acc_num}')
    print(f'**** 잔 고 ****   $ {self.__balance:,}')
    print('\n\n')

  @property
  def deposit_history(self):                                  # Q10
    print('-' * 10)
    for credit in self.__transac_info['입금내역']: print(credit)
    print('-' * 10)


  @property
  def withdraw_history(self):                                 # Q10
    print('-' * 10)
    for credit in self.__transac_info['출금내역']: print(credit)
    print('-' * 10)
```

```python
# Q8
my_acc1 = Account()
my_acc1.get_account_num

my_acc2 = Account()
my_acc2.get_account_num

my_acc3 = Account()
my_acc3.get_account_num

my_acc4 = Account()
my_acc4.get_account_num


bank_acc_list = [my_acc1, my_acc2, my_acc3, my_acc4]


# Q9
for customer_acc in bank_acc_list:
  customer_acc.balance
  for i in range(20):
    customer_acc.deposit(randint(1000000, 9999999))
  for i in range(20):
    customer_acc.withdraw(randint(200000, 1000000))
  if customer_acc.balance >= 1000000: customer_acc.display_info

  customer_acc.deposit_history
  customer_acc.withdraw_history
```
